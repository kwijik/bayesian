```python
# What if we have a 4sided globe ?
```

```python
sample = list("WLWWWLWLW")
```

```python
countW = sample.count('W')
```

```python
countL = sample.count('L')
```

```python
countW, countL
```

```python
n = len(sample)
```

```python
print(f"W = {countW}, L = {countL}, total obs = {n}\n")
```

```python
p = np.array([0.0, 0.25, 0.5, 0.75, 1.0])
```

```python
ways = (4 * p) ** countW * (4 * (1 - p)) ** countL
```

```python
prob = ways / ways.sum()
```

```python
prob
```

```python
print(" p     likelihood       posterior")
print("-" * 35)
for p_val, way, pr in zip(p, ways, prob):
    print(f"{p_val:4.2f}   {way:12.4g}    {pr:9.4f}")
```

```python
# Simulating uncertainty with Beta distribution
```

```python
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
alpha = 7
beta_param = 4
n_samples = 2000
```

```python
post_samples = beta.rvs(alpha, beta_param, size=n_samples)
```

```python
plt.hist(post_samples, bins=30, density=True)
```
