```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# simulating weight from height
```

```python
def sim_weight(H, b, seed=None):
    if seed is None:
        seed = np.random.normal(loc=10, scale=5.0, size=None)
    H = np.asarray(H)
    U = np.random.normal(0, seed, size=len(H))
    W = H*b + U
    return W
    
```

```python
#H = [160, 170, 180]
H = random_numbers = np.random.normal(loc=170, scale=20, size=90)

b = 0.6
sd = 5
```

```python
W =sim_weight(H, b, sd)
```
