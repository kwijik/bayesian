```python
def sim_globe(p=0.7, N=9):
    return np.random.choice(["W", "L"], size=N, p=[p,1-p], replace=True)
```

```python
sim_globe()
```

```python
[[sim_globe()] for _ in range(10)]
```

```python
#testing the simulation
```

```python
sim_globe(p=1, N=11)
```

```python
sum(x == "W" for x in sim_globe(p=0.5, N=int(1e4))) / 1e4
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
import numpy as np

successes = np.array([148, 212])

trials = np.array([9469, 7025])

stat, pval = proportions_ztest(successes, trials)

print(f"Z-stats: {stat:.4f}")
print(f"P-value: {pval:.10f}")

if pval < 0.05:
    print("stats significant")
else:
    print("not significant")
```

```python
successes = np.array([44, 68])

trials = np.array([9469, 7025])

stat, pval = proportions_ztest(successes, trials)

print(f"Z-stats: {stat:.4f}")
print(f"P-value: {pval:.10f}")

if pval < 0.05:
    print("stats significant")
else:
    print("not significant")
```
