```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# E [divorce | mariage , age] ; mu_i = alfa + beta_A * Ai + beta_M * Mi
```

```python
n = 20 
a = np.random.normal(loc=0, scale=10, size=n) 
bA = np.random.normal(loc=0, scale=10, size=n)
```

```python
plt.figure(figsize=(8, 6))
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel("Median age of marriage (st-d)")
plt.ylabel("Divorce rate (st-d)")

Aseq = np.linspace(-3, 3, 30)

for i in range(n):
    mu = a[i] + bA[i] * Aseq 
    plt.plot(Aseq, mu, color='red', linewidth=2)

#plt.show()
```
