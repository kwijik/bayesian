```python
# sex-effect male-female diff
```

```python
n = 100
```

```python
H_f = 150 + np.random.normal(0, 5, n)
W_f = 0.0 + 0.5 * H_f + np.random.normal(0, 5, n)
```

```python
H_m = 160 + np.random.normal(0, 5, n)
W_m = 0.0 + 0.6 * H_m + np.random.normal(0, 5, n)
```

```python
effect = W_m.mean() - W_f.mean() #total causal effect
```
